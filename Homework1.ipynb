```python
import cv2
import numpy as np
```

```python
def nothing(x):
    pass
```

```python
drawing=False#trueifmouseispressed
ix,iy=-1,-1
cx,cy=-1,-1
org_img = cv2.imread('mountain.jpg')
img = org_img.copy()
#mousecallbackfunction
def draw_circle(event,x,y,flags,param):
    global ix,iy,cx,cy,drawing,mode,img
    if event==cv2.EVENT_LBUTTONDOWN:
        if drawing == False:
            drawing = True
            ix, iy = x,y
            cx, cy = x, y
        else:
            drawing = False
            cx, cy = x, y
    elif event==cv2.EVENT_MOUSEMOVE:
        if drawing==True:
            cx,cy=x,y
```

```python
cv2.namedWindow('image')
cv2.setMouseCallback('image',draw_circle)
cv2.createTrackbar('value','image',0,255,nothing)
font = cv2.FONT_HERSHEY_SIMPLEX
while(1):
    img = org_img.copy()
    value = cv2.getTrackbarPos('value','image')
    if ix != -1:
        overlay = img.copy()
        cv2.rectangle(overlay,(ix,iy),(cx,cy),(0,0,value),-1)
        alpha = 0.5
        cv2.addWeighted(overlay, alpha, img, 1 - alpha, 0, img)
    cv2.putText(img,f"Mouse position : ({ix},{iy})-({cx},{cy})-({value})",(10,30), font, 0.7,(255,255,255),2,cv2.LINE_AA)
    cv2.imshow('image',img)
    k=cv2.waitKey(1)&0xFF
    if k== 27:
        break
cv2.destroyAllWindows()
```
