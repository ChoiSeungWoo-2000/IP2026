## List Comprehensions


```python
li = [3, 6, 2, 7]
out = []
for elem in li:
    print(out)
    out.append(elem*12) # 곱하기 넣으면 두배가능
print(out)
```

```python
[elem*12 for elem in li] # 위에껄 합해놓은게 이거임. 리스트 컴프레션  ★
```

```python
li = [('a',1), ('b',2), ('c',3), ('c',7)]  ## 중간고사임. 빈칸채우기 걍 통으로 외워라,.
[ n * 3 for (x,n) in li]
```

```python
li = [3, 6, 2, 7, 1, 9]
[elem* 2 for elem in li if elem>4]
```
