## Basic Operations on Images


```python
import numpy as np
import cv2
```

```python
img = cv2.imread('mountain.jpg')
```

```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
px = img[100,100]
print(px) #BGR 로 나온다. RGB아님 Open CV에서
```

```python
blue = img[100,100,0]
print(blue)  # 오로지 blue 값만
```

```python
img[100,100] = [255,255,255]
print(img[100,100])  #칼라값 수정하는법;
```

```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img[100:200,100:200] = [0,0,255] # 100에서 200까지 파란색으로 바꿈.
```

```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img[200:400,100:300,2] = 255
```

```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
