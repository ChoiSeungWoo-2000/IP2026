## Fuctions

```python
def myfun(x,y):
    return x * y 
myfun(3, 4)
```

```python
def myfun(x):
    return x * 3

def apply(q, x):
    return q(x)

apply(myfun, 7)
```

```python
apply(lambda z: z * 4, 7)  # 람다함수임.
```

```python
def lambda1(z):  #예약어라 에러뜸 그래서 1붙임.
    return z * 4
```

```python
apply(lambda1, 7)
```

```python
def myfun(b, c=3, d="hello"):
    return b+c
myfun(5,3,"hello")
myfun(5,3)
myfun(5)
```

```python
def myfun(a, b, c):
    return a-b
myfun(2,1,43)
myfun(c=43, b=1, a=2)
myfun(2, c=43, b=1)
```
