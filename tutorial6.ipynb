## For Loops


```python
#for <item> in <collection>: <statements>
for someChar in "Hello World":
    print (someChar)
```

```python
a = 1
b = 2
c = 3
d = 4
for (x, y) in [(a,1), (b,2), (c,3), (d,4)]:
    print (x)
for (x, y) in [(a,4), (b,3), (c,2), (d,1)]:
    print (y)
```

```python
for x in range(5):
    print(x)
```

```python
mylist = [10,100,100]
for i in range(len(mylist)):
    print (i, mylist[i])
```

```python
mylist = [5,15,155]
for (i, item) in enumerate(mylist):
    print (i, item)
```
