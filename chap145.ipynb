# Morphological Transformations

```python
import cv2
import numpy as np
```

```python
img = cv2.imread('charj.png',0)
kernel = np.ones((5,5),np.uint8)
erosion1 = cv2.erode(img,kernel,iterations = 1)
erosion2 = cv2.erode(img,kernel,iterations = 2)
erosion3 = cv2.erode(img,kernel,iterations = 3)
```

```python
cv2.imshow('img',img)
cv2.imshow('dst1',erosion1)
cv2.imshow('dst2',erosion2)
cv2.imshow('dst3',erosion3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('charj.png',0)
kernel = np.ones((5,5),np.uint8)
erosion1 = cv2.dilate(img,kernel,iterations = 1)
erosion2 = cv2.dilate(img,kernel,iterations = 2)
erosion3 = cv2.dilate(img,kernel,iterations = 3)
```

```python
cv2.imshow('img',img)
cv2.imshow('dst1',erosion1)
cv2.imshow('dst2',erosion2)
cv2.imshow('dst3',erosion3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('charj2.png',0)
kernel = np.ones((5,5),np.uint8)
erosion1 = cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel,iterations = 2)
```

```python
cv2.imshow('img',img)
cv2.imshow('dst1',erosion1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('charj3.png',0)
kernel = np.ones((5,5),np.uint8)
erosion1 = cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel,iterations = 2)
```

```python
cv2.imshow('img',img)
cv2.imshow('dst1',erosion1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
