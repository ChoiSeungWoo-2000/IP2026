## Geometric Transformations of Images



```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('opencv_logo.png',0)
rows,cols = img.shape
res1 = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_LINEAR)
res2 = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_AREA)
res3 = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
```

```python
cv2.namedWindow('res1')
cv2.namedWindow('res2')
cv2.namedWindow('res3')
cv2.resizeWindow("res1", rows*2, cols*2)
cv2.resizeWindow("res2", rows*2, cols*2)
cv2.resizeWindow("res3", rows*2, cols*2)
while(1):
    cv2.imshow('res1',img)
    cv2.imshow('res2',img)
    cv2.imshow('res3',img)
    if cv2.waitKey(20)&0xFF==27:
        break
cv2.destroyAllWindows()
```

## Translation

```python
rows,cols = img.shape
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
cv2.imshow('img',img)
cv2.imshow('dst',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Rotation

```python
M = cv2.getRotationMatrix2D((cols/2,rows/2),45,1)
dst = cv2.warpAffine(img,M,(cols,rows))
cv2.imshow('img',img)
cv2.imshow('dst',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Affine Transformation

```python
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

## Perspective Transformation

```python
img = cv2.imread('dsu1.jpg')   #중간고사에 나옴. 코드해석
rows,cols,ch = img.shape
pts1 = np.float32([[323,282],[760,310],[765,509],[322,534]])
pts2 = np.float32([[300,300],[800,300],[800,600],[300,600]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(1100,900))
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
